```python
import requests
import json
propertyLabelCache = {}

# this sparql asks for all the property names
sparql = """
	SELECT ?property ?propertyLabel WHERE {
	    ?property a wikibase:Property .
	    SERVICE wikibase:label {
	      bd:serviceParam wikibase:language "en" .
	   }
	 }
"""

headers = {
    'Accept': 'application/sparql-results+json',
}

params = {
	'query' : sparql
}

response = requests.get(
    'https://query.semlab.io/proxy/wdqs/bigdata/namespace/wdq/sparql',
    params=params,
    headers=headers,
)

data = response.json()

propertyLabelCache = {}

for result in data['results']['bindings']:

	propertyLabelCache[result['property']['value'].split('/entity/')[1]] = result['propertyLabel']['value'] + f" ({result['property']['value'].split('/entity/')[1]})"


# now ask for all the propertys being used on items that are a specific type and part of project
sparql = """
	SELECT ?p ?pLabel
	WHERE 
	{
	  
	  VALUES ?instaceOfTypes {wd:Q1 wd:Q18805} .
	  
	  ?item wdt:P1 ?instaceOfTypes .
	  ?item wdt:P11 wd:Q21217 .
	  ?item ?p ?o .
	  
	  
	  SERVICE wikibase:label { bd:serviceParam wikibase:language '[AUTO_LANGUAGE],en'. } # Helps get the label in your language, if not, then en language
	}
"""

headers = {
    'Accept': 'application/sparql-results+json',
}

params = {
	'query' : sparql
}

response = requests.get(
    'https://query.semlab.io/proxy/wdqs/bigdata/namespace/wdq/sparql',
    params=params,
    headers=headers,
)

data = response.json()

countLookup = {}

for result in data['results']['bindings']:


	if 'http://base.semlab.io/prop/' in result['p']['value']:
		if '/prop/P' in result['p']['value']:
			Pid = 'P' + result['p']['value'].split('/prop/P')[1]
			label = propertyLabelCache[Pid] 
			if label not in countLookup:
				countLookup[label] = 0

			countLookup[label] = countLookup[label] + 1

print(json.dumps({k: v for k, v in sorted(countLookup.items(), key=lambda item: item[1], reverse=True)},indent=2))

```
